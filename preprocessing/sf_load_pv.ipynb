```python
data_location = '../datasets/'
sf_load_df_raw = pd.read_csv(data_location +  'energy/SanFrancisco_Hospital.csv')
sf_pv_df_raw = pd.read_csv(data_location +  'energy/SanFrancisco_PV_GHI.csv')
```
